```
import numpy as np
import matplotlib.pyplot as plt

G=9.80665
t=0.0
h=0.01

v=float(input("velocity:"))
x=float(input("height"))
print("{:.7f} {:.7f} {:.7f}".format(t,v,x))
tlist=[t]
xlist=[x]
vlist=[v]

while x>=0:
  t+=h
  v+=G*h
  x-=v*h
  tlist.append(t)
  xlist.append(x)
  vlist.append(v)

plt.plot(tlist,xlist)
plt.show()
plt.plot(tlist,vlist)
plt.show()
```

```
F=1.5
G=9.80665

def retrofire(t,tf):
  if t>=tf:
    return -F*G
  else:
    return 0.0


t=0.0
h=0.01

v=float(input("velocity:"))
x0=float(input("height:"))
tf=float(input("time for getting fire:"))
x=x0
print("{:.7f} {:.7f} {:.7f}".format(t,v,x))

while (x>0) and (x<=x0):
  t+=h
  v+=(G+ retrofire(t,tf))*h
  x-=v*h
  print("{:.7f} {:.7f} {:.7f}".format(t,v,x))
```

```
import numpy as np
import matplotlib.pyplot as plt


F=1.5
G=9.80665

def retrofire(t,tf):
  if t>=tf:
    return -F*G
  else:
    return 0.0


t=0.0
h=0.01

v=float(input("velocity:"))
x0=float(input("height:"))
tf=float(input("time for getting fire:"))
x=x0
print("{:.7f} {:.7f} {:.7f}".format(t,v,x))

tlist=[t]
xlist=[x]

while (x>0) and (x<=x0):
  t+=h
  v+=(G+ retrofire(t,tf))*h
  x-=v*h
  tlist.append(t)
  xlist.append(x)

plt.plot(tlist,xlist)
plt.show()
```
