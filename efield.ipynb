```
import math

Q=(((0.0,0,0),10.0),((5.0,5.0),5.0))
TIMELIMIT=20.0
RLIMIT=0.1
H=0.01

t=0.0

vx=float(input("v0x :"))
vy=float(input("v0y :"))
x=float(input("x :"))
y=float(input("y :"))

print("{:.7f} {:.7f} {:.7f} {:.7f} {:.7f}".format(t,x,y,vx,vy))

while t<TIMELIMIT:
  t=t+H
  rmin=float("inf")
  for qi in Q:
    rx=qi[0][0]-x
    ry=qi[0][1]-y
    r=math.sqrt(rx**2+ry**2)
    if r<rmin:
      rmin=r
    vx +=((rx/r**3)*qi[1])*H
    vy +=((ry/r**3)*qi[1])*H
  x+=vx*H
  y+=vy*H
  print("{:.7f} {:.7f} {:.7f} {:.7f} {:.7f}".format(t,x,y,vx,vy))

  if rmin<RLIMIT:
    break
```

```
import math
import numpy as np
import matplotlib.pyplot as plt


Q=(((0.0,0,0),10.0),((5.0,5.0),5.0))
TIMELIMIT=20.0
RLIMIT=0.1
H=0.01

t=0.0

vx=float(input("v0x :"))
vy=float(input("v0y :"))
x=float(input("x :"))
y=float(input("y :"))

print("{:.7f} {:.7f} {:.7f} {:.7f} {:.7f}".format(t,x,y,vx,vy))

ylist=[y]
xlist=[x]

while t<TIMELIMIT:
  t=t+H
  rmin=float("inf")
  for qi in Q:
    rx=qi[0][0]-x
    ry=qi[0][1]-y
    r=math.sqrt(rx**2+ry**2)
    if r<rmin:
      rmin=r
    vx +=((rx/r**3)*qi[1])*H
    vy +=((ry/r**3)*qi[1])*H
  x+=vx*H
  y+=vy*H

  xlist.append(x)
  ylist.append(y)

  if rmin<RLIMIT:
    break

plt.plot(xlist,ylist)
plt.show()
```
